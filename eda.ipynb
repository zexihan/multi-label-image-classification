## Distribution of labels

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
#np.set_printoptions(threshold=np.nan)
```

```python
train_df = pd.read_csv('data/train_without_missing.csv')
```

```python
str_list = [train_df['labelId'][i] for i in range(train_df.shape[0])]
int_list = [str_list[i].split(' ') for i in range(train_df.shape[0])]
flat_list = [int(item) for sublist in int_list for item in sublist]
```

```python
len(set(flat_list))
```

```python
plt.figure(figsize=(20,10))
plt.hist(flat_list, bins=228)
plt.show()
```
